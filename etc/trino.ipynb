```python
import pandas as pd
```

```python
from trino.dbapi import connect
from trino.auth import CertificateAuthentication
import requests

http_session = requests.Session()
http_session.cert= ('example.com.crt', 'example.com.key')
http_session.verify= False # don't want to verify SSL certificates

conn = connect(
    user="trino",
    http_session= http_session,
    http_scheme="https",
    host='trino.example.com',
    port= 443,
    schema='autos',
    catalog='minio'
)

```

```python
df = pd.read_sql('''
  select * from minio.autos.cars
''', conn)
df
```
